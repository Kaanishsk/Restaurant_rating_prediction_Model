```python
X = df[['Average Cost for two','Has Table booking','Has Online delivery','Price range']]
```

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
print(X.columns)
```

```python
X["Has Table booking"] = labelencoder.fit_transform(X["Has Table booking"])
```

```python
labelencoder.classes_
```

```python
X["Has Online delivery"] = labelencoder.fit_transform(X["Has Online delivery"])
```

```python
X
```

```python
print(f"Type of X: {type(X)}")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
def modelresults(predictions):
    print("Mean absolute error {}".format(mean_absolute_error(y_test,predictions)))
    print("Root mean squared error {}".format(np.sqrt(mean_squared_error(y_test,predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression();
```

```python
lr.fit(X_train,y_train)
```

```python
prediction_from_lr = lr.predict(X_test)
```

```python
modelresults(prediction_from_lr)
```

```python
import joblib
joblib.dump(prediction_from_lr,"mlmodel.pkl")
```
